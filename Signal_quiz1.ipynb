```python
# [rySol]

from thinkdsp import Sinusoid
import numpy as np


class JMTriSignal(Sinusoid):
    
        
    def evaluate(self, ts):
    
        θ= ts * 2 * np.pi * self.freq + self.offset 

        θ= θ % (np.pi * 2) 

        x0= θ *(-2/np.pi) +1
        x1= θ *(+2/np.pi) -3

        x= x0 * (θ<np.pi) + x1 * (θ>=np.pi) 

        x= x * self.amp 

        return x

class JMSqrSignal(Sinusoid):
    
        
    def evaluate(self, ts):
    
        θ= ts * 2 * np.pi * self.freq + self.offset 

        θ= θ % (np.pi * 2) 

        x0= -1 
        x1= +1 

        x= x0 * (θ<np.pi) + x1 * (θ>=np.pi) ！

        x= x * self.amp 

        return x       

class JMSawtoothSignal(Sinusoid):
    """ry Represents a sawtooth signal."""
        
    def evaluate(self, ts):
    
        θ= ts * 2 * np.pi * self.freq + self.offset 

        θ= θ % (np.pi * 2) 

        x= -1 + θ/np.pi 
       
        x= x * self.amp 

        return x
```
